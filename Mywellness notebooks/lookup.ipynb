```python
import pandas as pd
```

```python
df = pd.read_json("indooractivities-44efd75f-2b2f-44e2-a293-ec6caa3f836d-e146f083-60d0-4e78-b26e-dfc5f7cd45c2.json")
df.head()
```

```python
len(df.user.unique()) == 1 and len(df.facility.unique()) == 1
```

```python
df = df.drop(columns=['facility','user'])
df.head()
```

```python
# parse dates
df['on'] = pd.to_datetime(df['on'])
df['createdOn'] = pd.to_datetime(df['createdOn'])
```

```python
def single_set_convert(input_dict,**kwarg):
    assert input_dict['pos'] == 0
    result = kwarg
    for entry in input_dict['pr']:
        result[entry['n']] = entry['v']
    return result

def multiple_sets_convert(multiple_dicts):
    return [single_set_convert(entry,order=index) for index,entry in enumerate(multiple_dicts)]

def convert_exercise(entry):
    assert 'pr' in entry.keys() and 'st' in entry.keys()
    return multiple_sets_convert(entry['st'])

df['performedData'] = df['performedData'].apply(convert_exercise)
```

```python
phId_to_practice_range = df.groupby('phId').apply(lambda x:(x['on'].min().date(),x['on'].max().date())).reset_index(name="date_range")
phId_to_practice_range.groupby('date_range').apply(lambda x:x['phId'].values)
```

```python
exercises_types = df.drop_duplicates(subset=["phId"], keep='first')
exercises_types['performedDataStoredValues']  = exercises_types['performedData'].apply(lambda x: str(sorted(x[0].keys())) if len(x) > 0 else None)
exercises_types['performedDataStoredValues'].value_counts()
```

```python
exercises_types['Isweights'] = exercises_types['performedDataStoredValues'].apply(lambda x:True if x and 'IsoWeight' in x else False)
weights_phId = exercises_types[exercises_types['Isweights']]['phId']
```

```python
weights_phId
```

```python
weights_df = df[df['phId'].isin(weights_phId.values)]

def max_lifted(sets):
    return max(map(lambda x:x['IsoWeight'],sets))

def total_lifted(sets):
    return sum(map(lambda x:x['IsoWeight'],sets))

def number_of_sets(sets):
    return len(sets)


weights_df["max_lifted"] = weights_df['performedData'].apply(max_lifted)
weights_df["total_lifted"] = weights_df['performedData'].apply(total_lifted)
weights_df["number_of_sets"] = weights_df['performedData'].apply(number_of_sets)
weights_df.head()
```

```python
weights_df.groupby('phId').plot(y='max_lifted',x='on',kind='line')
```
